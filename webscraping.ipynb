```python
import requests
Link="https://www.javatpoint.com/top-10-cricketers-in-the-world"
Link_text=requests.get(Link).text
print(Link_text)
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(Link_text,'lxml')
```

```python
print(soup.prettify())
```

```python
print(soup.title)
```

```python
print(soup.title.string)
```

```python
soup.a
```

```python
soup.find_all('a')
```

```python
all_table=soup.find_all('table')
```
