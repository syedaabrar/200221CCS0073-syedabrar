```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,5]
plt.plot(x,y)
plt.title("info")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[4,5,4,5]
colors=["black","hotpink","b","#4CAF50"]
plt.barh(x,y,color=colors)
plt.xlabel("X axis",fontsize=20)
plt.ylabel("Y axis",fontsize=20)
plt.title("Bar plot")
plt.show
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,5,6,7]
colors=["black","hotpink","b","#4CAF50","g"]
sizes=[200,100,150,40,15]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.8,marker="*",edgecolor="g")
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.show()
```

```python
y=[35,25,25,15]
mylabels = ["Apples","Bananas","Cherries","Dates"]
mycolors = ["black","hotpink","b","#4CAF50"]
plt.pie(y, labels = mylabels,colors = mycolors)
plt.show()
```

```python
import numpy as np
y= np.array([35, 25, 25, 15])
mylabels = ["Apples","Bananas","cherries","Dates"]
myexplode = [0.2, 0, 0, 0.5]

plt.pie(y, labels = mylabels, explode = myexplode,shadow=True)
plt.show()
```
